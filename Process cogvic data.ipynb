```python
buffer_size = 7
processed_data = []
for i, row in tqdm(df.iterrows()):
    # chrom = grch38_chrom_mapping[row["Chr"]] 
    chrom = "chr"+row["Chr"]
    ix = row["Position"]-1
    if row.type not in ["stopgain", "nonsynonymous SNV", "synonymous SNV"]:
        continue
    if len(row.Ref)>1:
        continue
        
    if "-" in row.Ref or "-" in row.Alt:
        continue
        
    # assert hg_19_genome[chrom][ix]==row.Ref
    # assert grch38_genome[chrom][ix]==row.Ref
    
    if not hg_19_genome[chrom][ix]==row.Ref:
        continue
    
    preceding_sequence = hg_19_genome[chrom][ix-buffer_size:ix]
    following_sequence = hg_19_genome[chrom][ix+1:ix+buffer_size+1]
    # preceding_sequence = grch38_genome[chrom][ix-buffer_size:ix]
    # following_sequence = grch38_genome[chrom][ix+1:ix+buffer_size+1]
    
    
    processed_data.append({"ReferenceSequence": str(preceding_sequence).upper()+row.Ref+str(following_sequence).upper(), 
                           "MutatedSequence": str(preceding_sequence).upper()+row.Alt+str(following_sequence).upper(),
                          "type": row.type, "Molecular_Consequence": row.Molecular_Consequence, "Gene": row.Gene, "disease": row.disease,
                           "dbSNP": row["dbSNP"]
                          })
    
processed_data = pd.DataFrame(processed_data)
processed_data
```

```python
processed_data.to_csv("processed_data/cogvig_data.csv", index=False)
```
